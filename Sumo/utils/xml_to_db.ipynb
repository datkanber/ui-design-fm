```python
import xml.etree.ElementTree as ET
import MySQLdb
```

```python
XML_FILE = "Points.xml"
```

```python
db_connection = MySQLdb.connect(host = "localhost", user="root", password = "1234", database = "fleetmanagementdb")
cursor = db_connection.cursor()
```

```python
xml = ET.parse(XML_FILE)
xml_root = xml.getroot()

id = ''
name = ''
X = ''
Y = ''
lat = ''
lon = ''
type = ''

for child in xml_root:
    if child.tag == 'Point':
        id = child.attrib['No']
        name = child.attrib['Name']
        X = child.attrib['X']
        Y = child.attrib['Y']
        lat = float(child.attrib['lat'])
        lon = float(child.attrib['lon'])
        type = child.attrib['Type']
        mySql_insert_query = """INSERT INTO nokta (p_id, enlem, boylam, name) 
                                VALUES (%s, %s, %s, %s) """
        record = (id, lat, lon, name)
        cursor.execute(mySql_insert_query, record)

    
```
